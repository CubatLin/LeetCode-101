```python
import collections

def subarraySum( nums, k):
    d = collections.defaultdict(int)
    d[0] = 1
    sum = 0; res = 0
    for i in range(len(nums)):
        sum += nums[i]
        print("sum:"+str(sum))
        if sum - k in d:
            print('get k:', str(sum - k))
            res += d[sum - k]
            print("res:"+str(res))
        d[sum] += 1
        print(f"d[sum = {sum}]:"+str(d[sum]))
        print('-------')
    return res
```

```python
nums = [1,1,1,2,1,-1,1,2,1]
k=3

subarraySum(nums,k)
```
