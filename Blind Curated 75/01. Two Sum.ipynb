```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        # 題目其實是找尋一個元素在某個list裡（因為第一個元素的index已經可以找到)
        # 因為要回傳位置，所以set不是好方法；而且set在一開始就會全包起來(記憶體會爆)
        # 用hashmap比較好, 他是動態新增記憶體
        # step: O(n)最多只走n步 / memory: O(n)最多新增n個元素
        index_hashmap = {}        
        for idx, num in enumerate(nums):
            if target - num in index_hashmap:
                return [idx, index_hashmap[target - num]]
        
            index_hashmap[num] = idx
```
