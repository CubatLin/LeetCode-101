Given a sorted array of distinct integers and a target value, return the index if the target is found. If not, return the index where it would be if it were inserted in order.

```  
Input: nums = [1,3,5,6], target = 5
Output: 2

Input: nums = [1,3,5,6], target = 7
Output: 4

Input: nums = [1,3,5,6], target = 0
Output: 0
```

##Super Easy
```  
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        for i in range(len(nums)):
            if nums[i]>=target: # 當target在開頭 & target=某數(等於) &target要被塞在某數之間(大於)   
                return i
        return len(nums)#如果都不是,就是放在最後一個位置
```            

## Binary Search 
* 只有四種可能結果, 其中三種return left,另外一種 當值有在array裡時, 才return mid
* O(log2(n))

````python
from IPython.display import Image
Image("bs.jpg", width=850)
````

``` 
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        l = 0 #left
        r = len(nums)-1 #bottom #左閉右閉
        while l<=r:
            mid = (l+r)//2
            if nums[mid]<target: l = mid+1 #加左邊讓l>r then break while loop
            elif nums[mid]>target :r = mid-1  #減右邊讓r<l then break while loop
            else: return mid
        return l
```
