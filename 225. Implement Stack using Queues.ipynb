#### 用Queues實現Stack
Implement a last in first out (LIFO) stack using only two queues. The implemented stack should support all the functions of a normal queue (push, top, pop, and empty).
https://leetcode.com/problems/implement-stack-using-queues/
* Implement the MyStack class:
 * void push(int x) Pushes element x to the top of the stack.
 * int pop() Removes the element on the top of the stack and returns it.
 * int top() Returns the element on the top of the stack.
 * boolean empty() Returns true if the stack is empty, false otherwise.

## 知識點:
1. 堆(Stack)是以底層容器完成其所有的工作，對外提供統一的接口，底層容器是可插拔的（也就是說我們可以控制使用哪種容器來實現Stack的功能）。


2. 堆的具體流程(這種容器沒有在遍歷所有元素的, 就是從最尾端開始拔, 雙向的話再考慮由頭開始拔或尾巴開始拔)



### Type 1: Use List as container

```
class MyStack:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.container=[] #Using list as container

    def push(self, x: int) -> None:
        """
        Push element x onto stack.
        """
        self.container+=[x]
        #self.container.append(x)

    def pop(self) -> int:
        """
        Removes the element on top of the stack and returns that element.
        """
        return self.container.pop()

    def top(self) -> int:
        """
        Get the top element. 
        """
        return self.container[-1]#top means last

    def empty(self) -> bool:
        """
        Returns whether the stack is empty.
        """
        return self.container==[]
       
# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()
```
